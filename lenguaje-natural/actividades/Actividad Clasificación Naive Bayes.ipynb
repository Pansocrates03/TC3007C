## Ejercicio 1
 Assume the following likelihoods for each word being part of a positive or  negative movie review, and equal prior probabilities for each class.



 

 What class will Naive bayes assign to the sentence ‚ÄúI always like foreign films.‚Äù?

```python
sentence = "I always like foreign films."

bank = {
    "I": {
        "positive": 0.09,
        "negative": 0.16,
    },
    "always": {
        "positive": 0.07,
        "negative": 0.06,
    },
    "like": {
        "positive": 0.29,
        "negative": 0.06,
    },
    "foreign": {
        "positive": 0.04,
        "negative": 0.15,
    },
    "films": {
        "positive": 0.08,
        "negative": 0.11,
    },
}

def sentiment_analysis(sentence, bank):
    words = sentence.split()
    total_positive = 0
    total_negative = 0
    for word in words:
        word_lower = word.lower().strip('.,!?;"\'()')
        if word_lower in bank:
            total_positive += bank[word_lower]["positive"]
            total_negative += bank[word_lower]["negative"]
    if total_positive > total_negative:
        return "Positive"
    elif total_negative > total_positive:
        return "Negative"
    else:
        return "Neutral"
    
result = sentiment_analysis(sentence, bank)
print(f"The sentiment of the sentence is: {result}")
```
